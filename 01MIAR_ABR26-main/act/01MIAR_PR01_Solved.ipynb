## 01 - FRange

- Crear una lista con números que comiencen por 30 y que termine en 36 haciendo que el paso entre cada número consecutivo sea de 0.4. Sin usar Numpy.

```python
# Test Use Case
expected_output = [30, 30.4, 30.8, 31.2, 31.6, 32.0, 32.4, 32.8, 33.2, 33.6, 34.0, 34.4, 34.8, 35.2, 35.6, 36.0]
output = frange(30, 36, 0.4)    # This calls your function named frange

print(expected_output)
print(output)

assert(expected_output == output)    # This will fail if the output is wrong
```

```python
# Your Solution Here!!!
import math

def frange(init, finish, step):
    numbers = []
    while(init <= finish or math.isclose(init,finish)):
        numbers.append(round(init,2))
        init += step
    return numbers

#frange(30, 36, 0.4)
```

```python
def frange(start, end, step):

    number_of_steps = int((end - start) / step) + 1

    output_list = []

    for i in range(0,number_of_steps):

        output_list.append(round(start + float(i)*step,2))

    return output_list

#frange(3, 3.6, 0.4)
```

```python
def frange(i,f,p):
    contador = (f-i)/p
    salida = [i]
    while contador > 0:
        salida.append(round(i+p,1))
        i += p
        contador = contador - 1
    return salida
```

## Recursive RLE

- RLE (Run Lenght Encoding) es una forma simple de compresión de datos, que consiste en contar cuantas veces consecutivas aparece un valor en una secuencia.  Por ejemplo, dada una cadena de caracteres 'dddddwwoooodddaaa', el algoritmo RLE lo transforma en 'd5w2o4d3a3', convirtiendo una cadena de 17 elementos en otra de 10 elementos que representan los mismos datos.
- Implementar RLE de manera iterativa para una cadena de caracteres, de manera que dado un string devuelva otro string, cuya secuencia sea el carácter seguido del número de repeticiones de dicho carácter.

```python
# Test Use Case
string = 'dddddwwoooodddaaa'

expected_output = 'd5w2o4d3a3'
output = rle(string)    # This calls your function named rle

print(expected_output)
print(output)

assert(expected_output == output)    # This will fail if the output is wrong
```

```python
# Your solution HERE !!!!
def rle(string):
    output = string[0]
    string = string[1:]
    cuenta = 1
    
    while(string):
        caracter = string[0]
        if(caracter == output[-1]):
            cuenta += 1
        else:
            output = output + str(cuenta) + caracter
            cuenta = 1
        string = string[1:]

    output += str(cuenta)

    return output
```

```python
def rle(s):
    if not s:
        return ''
    contar = 1
    while contar < len(s) and s[contar] == s[0]:
        contar += 1
    comprimido = s[0] + str(contar)
    return comprimido + rle(s[contar:])
```

```python
def rle(string):
    result = []
    count = 1
    car_prev = string[0]
    for char in string[1:]:
        if char == car_prev:
            count += 1
        else:
            result.append(f"{car_prev}{count}")
            car_prev = char
            count = 1
    result.append(f"{car_prev}{count}")
    return ''.join(result)
```
