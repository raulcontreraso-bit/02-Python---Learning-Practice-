# 01MIAR - Actividad Whitepapers

### Introducción:

- Para abordar un proyecto de Inteligencia Artificial es necesario disponer del conocimiento suficiente tanto a nivel teórico como técnico para garantizar el éxito en la consecución de los objetivos. La fase de investigación sobre la temática de un proyecto es indispensable para conocer nuevos enfoques o afianzar y dar sustento a las acciones que se llevarán a cabo en los trabajos. Los artículos científicos son la referencia más utilizada para investigar sobre un ámbito de conocimiento ya que, antes de su publicación, son aceptados en un proceso llamado “revisión por pares”.

### Objetivo

- Investigar sobre el estado actual del conocimiento sobre procesamiento de datos en trabajos relacionados con Inteligencia Artificial, en concreto, con el uso de Python y
librerías.

- El propósito de esta actividad es conocer e indagar en los contenidos definidos en la asignatura Python para la Inteligencia Artificial mediante artículos científicos, siendo
estos el marco del conocimiento a investigar.

### Metodología

- Buscar y seleccionar un artículo científico relacionado con la asignatura, es decir, Python o IA.
    - Google Académico (Scholar) es un buscador muy recomendado, aporta una primera idea sobre qué hay publicado y en qué revistas científicas.
    - También se acepatarán artículos publicados en conferencias (proceedings) o "preprints" en [arXiv](https://arxiv.org/) siempre y cuando tengan un formato académico. 
    - La fecha de publicación del artículo debe ser actual, no se aceptarán artículos previos al 2024.
- Crear un notebook de Python que servirá como entrega de la actividad (se entregará el fichero .ipynb). En esta entrega se debe:
    - Referenciar adecuadamente este artículo con su estilo de preferencia (IEEE, Harvard, Science).
    - Describir porqué has seleccionado este artículo (trabajo, estudios, hobbies, otros intereses...). Explicar que quieres aprender y su usabilidad.
    - Resumir el contenido del artículo.
    - Seleccionar un ejemplo de implementación con Python. El código puede ser extraído directamente del artículo.
    - Explicar el código implementado.
    - Concluir con un párrafo de reflexión. ¿Qué has aprendido?
- Se recomienda:
    - Formatear bien los apartados propuestos con celdas Markdown o de código según corresponda.
    - Mantener una estructura de código limpia, con comentarios oportunos y tomando en cuenta la secuencia en la que se ejecuta el código.
    - Utilizar un formato adecuado, si aplica, para referencias complementarias.
    - Sintetizar la información, aunque no hay límite de palabras.
    - Mantener una buena ortografía. 
  
**IMPORTANTE:**
- El uso de inteligencia artificial debe estar acorde a la normativa descrita en la guía docente.
- No está permitido compartir los resultados ni el código en ninguno de los foros.
